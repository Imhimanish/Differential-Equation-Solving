## **Euler's Method**

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
  return -y
def True_Solution(x1):
  return np.exp(-x1)
x = 0
y = 1
n = 500
h = 0.03
def Euler(f, x, y):
  X = []
  Y = []
  for i in range(n):
    x = x + i * h
    X.append(x)
    y = y + h * f(x, y)
    Y.append(y)
  return X, Y
X, Y = Euler(f, x, y)
x1=0
A=[]
B=[]
for i in range (n):
  x1=x1+i*h
  A.append(x1)
  b=True_Solution(x1)
  B.append(b)
 
plt.plot(X,Y,'b')
plt.show()
plt.plot(A,B,'g')
plt.show()
```
